```python
import tensorflow as tf
```

```python
x=tf.placeholder("float32",shape=[None,2])
y=tf.placeholder("float32",shape=[None])
W1 =tf.Variable([[1,1],[1,1]],dtype=tf.float32)
b1 =tf.Variable([1,1],dtype=tf.float32)
W2 =tf.Variable([[1,1],[1,1]],dtype=tf.float32)
b2 =tf.Variable([1,1],dtype=tf.float32)
loss = tf.norm((tf.multiply(W1,x)+b1)*(tf.multiply(W2,x)+b2)-y)
init =tf.global_variables_initializer()
sess=tf.Session()
sess.run(init)
```

```python
optimizer = tf.train.GradientDescentOptimizer(0.000001)
train = optimizer.minimize(loss)
```

```python
for i in range(50000):
    sess.run(train,feed_dict={x:[[1,2],[3,4],[2,5]],y:[4,16,25]})
    if i%10000 == 0:
        print(sess.run([loss],feed_dict={x:[[1,2],[3,4],[2,5]],y:[4,16,25]}))
```
